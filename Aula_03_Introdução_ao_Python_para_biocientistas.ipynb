# Função

### O que é uma Função?

Uma função é como uma pequena "máquina" que criamos para realizar uma tarefa específica.

Ela recebe dados de entrada (os parâmetros).

Executa uma série de passos.

Devolve um resultado (o retorno).

```python
# Calcular a área do quarto
largura_quarto = 3
altura_quarto = 4
area_quarto = largura_quarto * altura_quarto
print(f"A área do quarto é {area_quarto} m².")
```

```python
# Calcular a área da sala
largura_sala = 5
altura_sala = 6
area_sala = largura_sala * altura_sala
print(f"A área da sala é {area_sala} m².")
```

```python

def calcular_area(largura, altura):
    """Esta função calcula a área de um retângulo."""
    area = largura * altura
    # return -> palavra que informa o que a função deve devolver como resultado
    return area

area_do_quarto = calcular_area(3, 4)
area_da_sala = calcular_area(5, 6)

print(f"Usando a função, a área do quarto é {area_do_quarto} m².")
print(f"Usando a função, a área da sala é {area_da_sala} m².")

# Podemos até usá-la diretamente dentro de um print!
print(f"A área de uma garagem 10x8 é {calcular_area(10, 8)} m².")
```

```python
# 1. Importar as bibliotecas necessárias
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


# Definimos os intervalos para nossas duas variáveis independentes.
temperatura = np.arange(10, 70, 1)  # Eixo X: de 10 a 70 °C
ph = np.arange(4, 11, 0.25)         # Eixo Y: de pH 4 a 11

# Criamos a grade de coordenadas com todas as combinações de Temperatura e pH.
T, P = np.meshgrid(temperatura, ph)

# Agora, o mais importante: o modelo matemático para a atividade enzimática (Z).
# Vamos modelar o efeito de cada variável como uma curva de sino (Gaussiana),
# onde a atividade é máxima no valor "ótimo" e diminui ao se afastar dele.

# Definimos as condições ótimas para a nossa enzima simulada.
temp_otima = 37.0
ph_otimo = 7.4

# O efeito da temperatura (curva gaussiana em torno da temp_otima)
efeito_temp = np.exp(-((T - temp_otima)**2) / (2 * 10**2))

# O efeito do pH (curva gaussiana em torno do ph_otimo)
efeito_ph = np.exp(-((P - ph_otimo)**2) / (2 * 1.5**2))

# A atividade enzimática total (Z) é o produto dos dois efeitos.
# A atividade só será 100% quando ambos estiverem no ótimo.
Z_atividade = efeito_temp * efeito_ph

# --- 3. CRIAÇÃO DO GRÁFICO 3D ---

# Criamos a figura e o eixo 3D, como no exemplo anterior.
fig = plt.figure(figsize=(11, 8))
ax = fig.add_subplot(111, projection='3d')

# Desenhamos a superfície, usando um mapa de cores diferente, como 'plasma'.
superficie = ax.plot_surface(T, P, Z_atividade, cmap='plasma')

# --- 4. ADICIONAR TÍTULOS E RÓTULOS ---
ax.set_title('Efeito da Temperatura e pH na Atividade Enzimática', fontsize=16)
ax.set_xlabel('Temperatura (°C)')
ax.set_ylabel('pH')
ax.set_zlabel('Atividade Relativa')

# Adiciona a barra de cores para referência
fig.colorbar(superficie, shrink=0.6)

# --- 5. EXIBIR O GRÁFICO ---
plt.show()
```

```python
# 1. Importar todas as bibliotecas necessárias
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from scipy.interpolate import griddata # A nova ferramenta para interpolação!

# --- 2. PREPARAÇÃO DOS DADOS EXPERIMENTAIS (SIMULADOS) ---

np.random.seed(42) # Para resultados reprodutíveis
temp_exp = 20 + 50 * np.random.rand(50) # 100 temperaturas entre 10 e 70
ph_exp = 4 + 6 * np.random.rand(50)     # 100 pHs entre 4 e 11

# A atividade (Z) será calculada pela nossa função modelo + um ruído aleatório
# para simular um erro experimental.
temp_otima = 37.0
ph_otimo = 7.4
efeito_temp = np.exp(-((temp_exp - temp_otima)**2) / (2 * 10**2))
efeito_ph = np.exp(-((ph_exp - ph_otimo)**2) / (2 * 1.5**2))
ruido = 0.05 * np.random.randn(50) # Pequeno erro aleatório
atividade_exp = (efeito_temp * efeito_ph) + ruido

# --- 3. INTERPOLAÇÃO: De Pontos Esparsos para uma Grade Contínua ---

# Primeiro, criamos a grade regular onde queremos "desenhar" a superfície.
# Esta grade terá mais pontos (20x20) do que nossos dados experimentais.
grid_temp = np.linspace(20, 70, 20) # 20 pontos de temperatura
grid_ph = np.linspace(4, 10, 20)   # 20 pontos de pH
T_grid, P_grid = np.meshgrid(grid_temp, grid_ph)

# Agora, a mágica da interpolação com griddata!
# Pedimos à SciPy para "adivinhar" os valores de atividade (Z) para cada
# ponto da nossa nova grade, baseando-se nos dados experimentais.
Z_grid = griddata(
    (temp_exp, ph_exp), # Coordenadas X, Y dos nossos dados originais
    atividade_exp,      # Valores Z dos nossos dados originais
    (T_grid, P_grid),   # A nova grade onde queremos os valores
    method='cubic'      # Método de interpolação ('linear' ou 'cubic')
)


# --- 4. CRIAÇÃO DO GRÁFICO 3D ---
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Desenhamos a superfície interpolada
superficie = ax.plot_surface(T_grid, P_grid, Z_grid, cmap='plasma', alpha=0.8)

# Isso mostra como a superfície foi construída a partir dos nossos dados.
ax.scatter(temp_exp, ph_exp, atividade_exp, color='red', s=20, label='Pontos Experimentais')

# --- 5. TÍTULOS, RÓTULOS E LEGENDA ---
ax.set_title('Superfície de Atividade Enzimática Interpolada de Dados Experimentais')
ax.set_xlabel('Temperatura (°C)')
ax.set_ylabel('pH')
ax.set_zlabel('Atividade Relativa')
ax.legend()
fig.colorbar(superficie, shrink=0.5)

# --- 6. EXIBIR O GRÁFICO ---
plt.show()
```

```python
# 1. Importar a biblioteca
import matplotlib.pyplot as plt

# Nossos grupos/categorias para o eixo X
grupos = ['Controle', 'Droga A', 'Droga B', 'Droga C']

# Os resultados médios de viabilidade para cada grupo (alturas das barras)
viabilidade_media = [100, 75, 45, 82] # Em porcentagem

# O desvio padrão ou erro padrão de cada medida (tamanho das barras de erro)
# Isso representa a variabilidade entre as réplicas do experimento.
erro_padrao = [5, 8, 4, 6]


# --- 3. CRIAÇÃO DO GRÁFICO DE BARRAS ---

plt.figure(figsize=(8, 6)) # Ajusta o tamanho da figura

# A função principal é plt.bar()
# Parâmetros importantes:
#   - x: As categorias no eixo X.
#   - height: Os valores médios (altura das barras) no eixo Y.
#   - yerr: Os valores de erro para criar as barras de erro.
#   - capsize: O tamanho das "travessas" nas pontas das barras de erro.
#   - color: Define as cores das barras.
plt.bar(
    grupos,
    viabilidade_media,
    yerr=erro_padrao,
    capsize=5,
    color=['gray', 'skyblue', 'salmon', 'lightgreen']
)

# --- 4. ADICIONAR TÍTULOS E RÓTULOS ---
plt.title('Efeito de Diferentes Drogas na Viabilidade Celular', fontsize=16)
plt.ylabel('Viabilidade Celular (%)')
# plt.xlabel('Grupos de Tratamento') # Opcional, pois os nomes dos grupos já são claros

# Ajusta o limite do eixo Y para dar um espaço extra no topo
plt.ylim(0, 120)


# --- 5. EXIBIR O GRÁFICO ---
plt.show()
```

```python
# 1. Importar as bibliotecas
import matplotlib.pyplot as plt
import numpy as np

# --- 2. PREPARAÇÃO DOS DADOS (UMA MATRIZ SIMPLES) ---

# Nossos genes e amostras
genes = ['GENE A', 'GENE B', 'GENE C', 'GENE D', 'GENE E']
amostras = ['Paciente 1', 'Paciente 2', 'Paciente 3', 'Paciente 4']

# Nossa matriz de dados de expressão (log2 fold change)
# Valores positivos (vermelho): gene mais expresso (up-regulated)
# Valores negativos (azul): gene menos expresso (down-regulated)
dados_expressao = np.array([
    [ 2.5,  1.0, -1.5, -0.5],  # Dados para o GENE A
    [-2.0, -1.0,  1.8,  1.2],  # Dados para o GENE B
    [ 1.8,  1.2, -0.8, -1.0],  # Dados para o GENE C
    [-1.5, -0.5,  2.5,  0.8],  # Dados para o GENE D
    [ 1.0,  0.5, -1.2, -2.5]   # Dados para o GENE E
])

# --- 3. CRIAÇÃO DO HEATMAP ---

fig, ax = plt.subplots(figsize=(8, 6))

# A função principal para o heatmap é a imshow()
# cmap='coolwarm' é um mapa de cores divergente, ótimo para dados positivos/negativos
im = ax.imshow(dados_expressao, cmap='coolwarm')

# --- 4. ADICIONAR RÓTULOS, TÍTULO E LEGENDA DE CORES ---

# Adicionar a barra de cores (colorbar) para sabermos o que cada cor significa
cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel('Log2 Fold Change', rotation=-90, va="bottom")

# Configurar os rótulos dos eixos X e Y
# Queremos mostrar os nomes dos genes e das amostras, não os números dos pixels.
ax.set_xticks(np.arange(len(amostras)))
ax.set_yticks(np.arange(len(genes)))

# E então colocar os nomes nesses locais
ax.set_xticklabels(amostras)
ax.set_yticklabels(genes)

# Rotacionar os rótulos do eixo X se eles forem longos
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

ax.set_title("Heatmap de Expressão Gênica")

# Para evitar que os rótulos sejam cortados na imagem final
fig.tight_layout()

# --- 5. EXIBIR O GRÁFICO ---
plt.show()
```
